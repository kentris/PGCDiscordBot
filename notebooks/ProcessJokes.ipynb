```python
from collections import Counter
import os
import pickle
import random
import time
```

```python
with open(os.path.join("..", "data", "dadjokes.txt")) as file:
    dadjokes = file.readlines()
```

```python
def process_dad_joke(dadjoke):
    msg1 = ""
    msg2 = ""
    if '? ' in dadjoke:
        msgs = dadjoke.split("? ", 1)
        msg1 = msgs[0] + '?'
        msg2 = msgs[1]
    elif '! ' in dadjoke:
        msgs = dadjoke.split("! ", 1)
        msg1 = msgs[0] + '!'
        msg2 = msgs[1]
    elif Counter(dadjoke)['.'] > 1:
        msgs = dadjoke.split(". ", 1)
        if len(msgs) > 1:
            msg1 = msgs[0] + '.'
            msg2 = msgs[1]
        else:
            msg1 = dadjoke
    else:
        msg1 = dadjoke
        
    return msg1, msg2
```

```python
for i, d in enumerate(dadjokes):
    if not '?' in d:
        m1, m2 = process_dad_joke(d)
        print(m1)
        time.sleep(2)
        if m2:
            print(m2)
```
